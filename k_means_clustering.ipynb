```
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
a= 10
sa = str(a)
X,y = make_moons(n_samples=200, noise=0.5, random_state=0)
kmeans = KMeans(n_clusters=a)
kmeans.fit(X)
y_pred = kmeans.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=60, edgecolors='k')
title = 'n_cluster='+sa
plt.title(title)
```

```
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],marker='^',s=100,linewidths=2,edgecolors='k')
plt.title(title)
plt.xlabel("v 0")
plt.ylabel("v 1")
```

```
import random
 
alist=[]                          # 뽑은 a를 넣어 중복 방지해주는 리스트         
for i in range(5):
  a = random.randint(1,15)       
  while a in alist :              # a가 이미 뽑은 리스트에 있을 때까지 다시 뽑자
    a = random.randint(1,30)
  alist.append(a) # 새로운 a 값을 리스트에 추가
 
print(alist)
```

```
from sklearn.datasets import make_blobs

def clustering(i,clu_std):

  a= i
  sa = str(a)
  X_v, Y_v = make_blobs(n_samples=200, cluster_std=clu_std, random_state=170)
  y_pred = KMeans(n_clusters=i, random_state=0).fit_predict(X_v)
  plt.scatter(X_v[:, 0], X_v[:, 1],c=y_pred, s=60, edgecolors='k')
  title = 'n_cluster='+sa
  plt.title(title)
      

  plt.show()
for i in alist:
  clustering(i,[2.5,7.5,0.5])
```
